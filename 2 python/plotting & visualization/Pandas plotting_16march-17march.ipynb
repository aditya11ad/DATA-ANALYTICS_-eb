```python
my_dict={
	'NAME':['Ravi','Raju','Alex','Ron','Geek','Kim'],
	'MARK':[20,30,40,30,40,50]
	}
df = pd.DataFrame(data=my_dict)
df
```

```python
df.plot.line(title="Std Mark");
```

```python
df.plot.line( title="Student Mark",figsize=(6,3),x='NAME',y='MARK',grid=True,legend=True,style='--')
```

```python
my_dict={
'NAME':['Ravi','Raju','Alex',
        'Ron','Geek','Kim'],
'MARK':[20,30,40,30,40,50]
}
df = pd.DataFrame(data=my_dict)
df.plot.bar(title="Std Mark",
            x='NAME')
```

```python
df.plot.bar(title="color",x='NAME',
            y='MARK',figsize=(5,3),
            color=(.9,.4,.6))
```

```python
df.plot.bar(title="color",x='NAME',
            y='MARK',figsize=(5,3),
            color=[(.9,.4,.6),(.1,.3,.8)])
```

```python
my_dict={
'NAME':['Ravi','Raju','Alex',
        'Ron','Geek','Kim'],
'MARK':[20,30,40,30,40,50],
'ENGLISH':[30,20,30,40,20,40]    
}
df = pd.DataFrame(data=my_dict)
```

```python
df.plot.bar(title="stacked=True",x='NAME',figsize=(5,3),stacked=False)
```

```python
df.plot.bar(title="stacked=True",x='NAME',figsize=(5,3),stacked=True)
```

```python
df.plot.barh(title="rot=180",x='NAME',y='MARK',
        figsize=(5,3),rot=45)
```

```python
my_dict={
'NAME':['Ravi','Raju','Alex','Ronn'],
'MATH':[30,40,50,50]
}
df = pd.DataFrame(data=my_dict)
```

```python
df
```

```python
df.plot.pie(title="Std Mark",y='MATH')
```

```python
df.plot.pie(title="Std Mark",y='MATH',labels=['Ravi','Raju','Alex','Ronn'],legend=False)
```

```python
df['MATH'].plot.pie(title="Std Mark",labels=['Ravi','Raju','Alex','Ronn'])
```

```python
df.plot.pie(title="Std Mark",y='MATH',labels=['Ravi','Raju','Alex','Ronn'],legend=False,explode=(.1,0,0,0))
```

```python
df.plot.pie(title="Std Mark",y='MATH',labels=['Ravi','Raju','Alex','Ronn'],legend=False,explode=(.3,0,0,0),shadow=True,fontsize=15)
```

```python
my_plot = df.plot.pie(title="Std Mark",y='MATH',labels=['Ravi','Raju','Alex','Ronn'],legend=False,explode=(.3,0,0,0),autopct='%.1f%%')
```

```python
fig = my_plot.get_figure()
```

```python
fig.savefig('my_saved_fig')
```
