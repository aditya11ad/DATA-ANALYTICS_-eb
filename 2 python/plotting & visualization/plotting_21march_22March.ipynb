```python
plt.plot(x, np.sin(x - 0), color='red') # specify color by name

plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)

plt.plot(x, np.sin(x - 2), color='.5') # Grayscale between 0 and 1

plt.plot(x, np.sin(x - 3), color='#FFFF00') # Hex code (RRGGBB from 00 to FF)

plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1

plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')

plt.plot(x, x + 1, linestyle='dashed')

plt.plot(x, x + 2, linestyle='dashdot')

plt.plot(x, x + 3, linestyle='dotted');
```

```python
plt.plot(x, x + 4, linestyle='-') # solid

plt.plot(x, x + 5, linestyle='--') # dashed

plt.plot(x, x + 6, linestyle='-.') # dashdot

plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
```

```python
x = np.linspace(0,4*3.14,50)
plt.figure(figsize=(9,3))
plt.subplot(2,1,1)
plt.plot(x,np.sin(x),':')
plt.subplot(2,1,2)
plt.plot(x,np.cos(x),'--')
```

#### subplots:

```python
f ,a =plt.subplots(2,figsize=(7,3))

a[0].plot(x,np.sin(x),':')
a[1].plot(x,np.cos(x),'--');
```

```python
f ,a =plt.subplots(2,figsize=(7,3),sharex=True)

a[0].plot(x,np.sin(x),':')
a[1].plot(x,np.cos(x),'--');
```

```python
f ,a =plt.subplots(1,2,figsize=(5,2),sharey=True)

a[0].plot(x,np.sin(x),':')
a[1].plot(x,np.cos(x),'--')
a[0].set_title('sin(x)')
a[1].set_title('cos(x)');
```

#### adding the axes in fig:

```python
x = np.linspace(0,4*3.14,50)
ax = plt.axes()
ax.plot(x,np.sin(x),':')
ax.set(title='sin(x)',xlabel='x',ylabel='sin(x)',xlim=(0,10),ylim=(-1,1));
```

```python
fig = plt.figure()
ax =fig.add_axes([0,0,1,1])
ax.plot(x,np.sin(x),':')
ax.set(title='sin(x)',xlabel='x',ylabel='sin(x)',xlim=(0,10),ylim=(-1,4))
ax2 =fig.add_axes([.6,.6,.4,.4])
ax2.plot(x,np.cos(x),'--')
ax2.set(title='cos(x)',xlabel='x',ylabel='cos(x)');
```

```python
x = np.random.normal(25,2,1000)
y = np.random.normal(30,3,1000)
z = np.random.normal(40,3,1000)
```

```python
plt.hist(x,alpha=.5,bins=40,label='X')
plt.hist(y,alpha=.5,bins=40,label='Y')
plt.hist(z,alpha=.5,bins=40,label='Z')
plt.legend();
```

```python
x,y= np.random.multivariate_normal([25,40],[[1,1],[2,1]],1000).T

plt.hist2d(x,y,bins=30,cmap='Blues')
plt.colorbar();
```
