#### outlier Detection, with the help of boxplots:

```python
for i in l:
    f,a=plt.subplots(1,2,figsize=(9,2))
    sns.histplot(x=i,data=df,ax=a[0])
    sns.boxplot(x=i,data=df,ax=a[1])
    plt.show()
```
